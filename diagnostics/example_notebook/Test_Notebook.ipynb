```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
config = {
    "dora_id": "odiv-2",
    "pathPP": "/path/to/some/experiment"
}
```

```python
for k, v in config.items():
    config[k] = (
        os.environ[f"MAR_{k.upper()}"]
        if f"MAR_{k.upper()}" in os.environ.keys()
        else v
    )
```

```python
print(config)
```

```python
x = np.linspace(1,100)
```

```python
fig1 = plt.figure(label="first_figure")
ax = plt.subplot(1,1,1)
ax.plot(x,x*2)
ax.text(0.02,0.95,config["pathPP"],ha="left",transform=ax.transAxes)
ax.text(0.02,0.90,config["dora_id"],ha="left",transform=ax.transAxes)
```

```python
fig2 = plt.figure(label="second_figure")
ax = plt.subplot(1,1,1)
ax.plot(x,np.sqrt(x))
```
